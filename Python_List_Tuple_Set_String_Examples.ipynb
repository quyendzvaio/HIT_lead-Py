```python
s = "Python c∆° b·∫£n"

print(s[0:6])   # "Python" ‚Äì t·ª´ index 0 ƒë·∫øn 5
print(s[:6])    # gi·ªëng tr√™n
print(s[7:])    # "c∆° b·∫£n" ‚Äì t·ª´ index 7 ƒë·∫øn h·∫øt
print(s[:])     # sao ch√©p to√†n b·ªô chu·ªói
```

```python
text = "abcdefg"

print(text[::2])   # "aceg" ‚Äì l·∫•y c√°ch 2 k√Ω t·ª±
print(text[::-1])  # "gfedcba" ‚Äì ƒë·∫£o ng∆∞·ª£c chu·ªói
print(text[1:6:2]) # "bdf"
```
