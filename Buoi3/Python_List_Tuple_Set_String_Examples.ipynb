üß© B√†i 2 ‚Äì X·ª≠ l√Ω chu·ªói (String)

Y√™u c·∫ßu:
Vi·∫øt ch∆∞∆°ng tr√¨nh nh·∫≠n v√†o m·ªôt chu·ªói, sau ƒë√≥:

ƒê·∫øm s·ªë l∆∞·ª£ng nguy√™n √¢m v√† ph·ª• √¢m trong chu·ªói.

In ra chu·ªói ƒë·∫£o ng∆∞·ª£c.

Ki·ªÉm tra xem chu·ªói c√≥ ph·∫£i l√† palindrome (ƒë·ªçc xu√¥i ng∆∞·ª£c nh∆∞ nhau) kh√¥ng.

```python
#lam bai vao day
strings = input("nhap chuoi ky tu:")
nguyen_am = "ueoai"
strings_na = sum(1 for char in strings if char in nguyen_am)
strings_pa = len(strings) - strings_na
print(f"so ki tu nguyen am la : {strings_na}")
print(f"so ki tu phu am la : {strings_pa}")
daonguoc = strings[::-1]
print(daonguoc)
if daonguoc == strings:
    print("chuoi dao nguoc la chuoi doi xung")
else:
    print("chuoi dao nguoc khong phai la chuoi doi xung")
```
